### Different Techniques to Create Tensors :

````python
# zeros and ones tensors
zeroes = torch.zeros(3,4)
ones = torch.ones(3,4)
print("zeros :" , zeroes)
print("ones :", ones)
````

````python
#Reproducibility : manual seed
# if you want the constant values every run , set the seed
torch.manual_seed(42)
a = torch.rand(2,2)
torch.manual_seed(42)
b = torch.rand(2,2)

print("a==b?",torch.equal(a,b))
print(a)
print(b)
````

````python
# *_like — same shape as another tensor
template= torch.rand(3,5)

z = torch.zeros_like(template)
o  = torch.ones_like(template)
r = torch.rand_like(template)

print(template.shape)
print(z.shape)
print("r:", r)
print("o:", o)
print("z:", z)
````
